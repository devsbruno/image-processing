```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image
import io

# Funo para converter imagem para escala de cinza
def to_grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Funo para converter imagem para binria
def to_binary(image, threshold=127):
    _, binary = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)
    return binary

# Upload de imagem
uploaded = files.upload()

for filename in uploaded.keys():
    image = Image.open(filename)
    image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

    # Converter para escala de cinza
    grayscale = to_grayscale(image_cv)

    # Converter para binria
    binary = to_binary(grayscale)

    # Exibir imagens
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB))
    axes[0].set_title("Imagem Colorida")
    axes[0].axis("off")

    axes[1].imshow(grayscale, cmap='gray')
    axes[1].set_title("Imagem em Tons de Cinza")
    axes[1].axis("off")

    axes[2].imshow(binary, cmap='gray')
    axes[2].set_title("Imagem Binria")
    axes[2].axis("off")

    plt.show()
```
